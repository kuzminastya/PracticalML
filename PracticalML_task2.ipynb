```python
import keras
import numpy as np
from sklearn.svm import LinearSVC
from keras.datasets import mnist
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix 
```

Загрузка набора данных

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

Подготовка данных: линеаризуем изображение, от 0 до 1

```python
X_train_scaled = np.zeros((60000, 28*28))
for i in np.arange(X_train.shape[0]):
    X_train_scaled[i] = np.reshape(X_train[i]/255, 28*28)
    
X_test_scaled = np.zeros((10000, 28*28))
for i in np.arange(X_test.shape[0]):
    X_test_scaled[i] = np.reshape(X_test[i]/255, 28*28)
```

Обучаем модель

```python
model = LinearSVC()
model.fit(X_train_scaled, y_train)
```

Рассчитываем confusion matrix

```python
plot_confusion_matrix(model, X_test_scaled, y_test)
```

```python
model.score(X_test_scaled, y_test)
```

Проверка 5 случайных образцов

```python
y_predict = model.predict(X_test_scaled)
for i in np.arange(5):
    index = np.random.randint(0,9999)
    print(y_test[index] == y_predict[index])
```
